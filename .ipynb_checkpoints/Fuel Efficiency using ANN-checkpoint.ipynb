```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import load_model
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
import sklearn
import numpy as np
from sklearn.metrics import (mean_squared_error, mean_absolute_error,r2_score)
import pickle
```

```python
df = pd.read_csv("fuel_mpg.csv", na_values='?')
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['car name'].unique()
```

# Drop Car Name 

```python
df.drop(['car name'],axis=1,inplace=True)
```

# Splitting in Train and Test

```python
X=df.drop("mpg",axis=1)
```

```python
Y=df['mpg']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train_std=sc.fit_transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

<h1 style="color:brown">saving SC transform</h1>

```python
pickle.dump(sc, open('models/scaler.pkl', 'wb'))
```

```python
Y_train
```

```python
Y_test
```

```python
X_train.keys()
```

```python
len(X_train.keys())
```

```python
# Initialising the ANN
model = Sequential()

# Adding the input layer and the first hidden layer
model.add(Dense(units = 64,activation='relu',input_shape=[len(X_train.keys())]))
# Adding the second hidden layer
model.add(Dense(units = 64,activation='relu'))
model.add(Dense(units = 64,activation='relu'))
model.add(Dense(units = 64,activation='relu'))
model.add(Dense(units=1))
# Adding the output layer

# Compiling the ANN
model.compile(loss='mean_squared_error',
                optimizer='Adam',
                metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
model.summary()
```

<h1 style="color:brown">Training with 100 Epchs</h1>

```python
# Fitting the ANN to the Training set
#model_history=model.fit(X_train_std, Y_train, batch_size = 100,epochs = 1000)
model_history = model.fit(
    X_train_std,
    Y_train,
    batch_size=16,
    epochs=100
)
```

```python
print(model_history.history.keys())
```

```python
model_history = model.fit(
    X_train_std,
    Y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2
)
```

```python
print(model_history.history.keys())
```

```python
import matplotlib.pyplot as plt

plt.plot(model_history.history['loss'], label='Train Loss')
plt.plot(model_history.history['val_loss'], label='Val Loss')
plt.legend()
plt.show()
```

```python
model.save("models/model.h5")
```

```python
X_test_std
```

```python
prediction=model.predict(X_test_std)
```

```python
prediction
```

```python
rmse=np.sqrt(mean_squared_error(Y_test,prediction))

print('RMSE: ',rmse)

print('R_square:', r2_score(Y_test,prediction))
```

```python
plt.scatter(Y_test,prediction)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])
```
